```python
from neuralprophet import NeuralProphet
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import pickle
```

```python
df=pd.read_csv('Mumbai_1990_2022_Santacruz.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['time'] = pd.to_datetime(df['time'], format='%d-%m-%Y')
```

```python
df.dtypes
```

```python
plt.plot(df['time'],df['tmax'])
```

```python
data=df[['time','tmax']]
data.dropna(inplace=True)
data.columns=['ds','y']
data.head()
```

**Training the model**

```python
model=NeuralProphet()  #untrained neuralprophet model
model.fit(data,freq='D',epochs=1000)
```

3.FORECASTING

```python
future=model.make_future_dataframe(data,periods=900)
forecast=model.predict(future)
forecast.head()
```

```python
forecast.tail()
```

```python
forecast
```

```python
plot1=model.plot(forecast)
```

```python
plot2=model.plot_components(forecast)
plt.show()
```

```python
forecast
```

```python
with open('forecast_model.pkl',"wb")as f:
    pickle.dump(model,f)
```
